```python
import pandas as pd
from  google.colab import files

up = files.upload()
```

```python
import io

df = pd.read_csv(io.BytesIO(up['data_raw_uber.csv']))
```

```python
df.head()
```

```python
df['pickup_datetime'] =  pd.to_datetime(df['tpep_pickup_datetime'])
df['drop_datetime'] =  pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
df.drop('tpep_pickup_datetime',axis=1,inplace=True)
```

```python
df.drop('tpep_dropoff_datetime',axis=1,inplace=True)
```

```python
datetime_dim = df[['pickup_datetime','drop_datetime']].reset_index(drop=True)
datetime_dim['pickup_datetime'] = datetime_dim['pickup_datetime']
datetime_dim['pick_hour'] = datetime_dim['pickup_datetime'].dt.hour
datetime_dim['pick_day'] = datetime_dim['pickup_datetime'].dt.day
datetime_dim['pick_month'] = datetime_dim['pickup_datetime'].dt.month
datetime_dim['pick_year'] = datetime_dim['pickup_datetime'].dt.year
datetime_dim['pick_weekday'] = datetime_dim['pickup_datetime'].dt.weekday

datetime_dim['drop_datetime'] = datetime_dim['drop_datetime']
datetime_dim['drop_hour'] = datetime_dim['drop_datetime'].dt.hour
datetime_dim['drop_day'] = datetime_dim['drop_datetime'].dt.day
datetime_dim['drop_month'] = datetime_dim['drop_datetime'].dt.month
datetime_dim['drop_year'] = datetime_dim['drop_datetime'].dt.year
datetime_dim['drop_weekday'] = datetime_dim['drop_datetime'].dt.weekday


datetime_dim['datetime_id'] = datetime_dim.index

# datetime_dim = datetime_dim.rename(columns={'tpep_pickup_datetime': 'datetime_id'}).reset_index(drop=True)
datetime_dim = datetime_dim[['datetime_id', 'pickup_datetime', 'pick_hour', 'pick_day', 'pick_month', 'pick_year', 'pick_weekday',
                             'drop_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
#
datetime_dim.head()
```

```python
pickup_location_dm = df[['pickup_longitude','pickup_latitude']].reset_index(drop=True)
pickup_location_dm['pickup_location_dm_index'] = pickup_location_dm.index

```

```python
pickup_location_dm = pickup_location_dm[['pickup_location_dm_index','pickup_longitude','pickup_latitude']]
pickup_location_dm.head()
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
payment_type_dm = df[['payment_type']].reset_index(drop=True)
payment_type_dm['payment_type_id'] = pickup_location_dm.index
payment_type_dm['payment_type_name'] = payment_type_dm['payment_type'].map(payment_type_name)
payment_type_dm = payment_type_dm[['payment_type_id','payment_type','payment_type_name']]
```

```python
payment_type_dm.head()
```

```python
drop_location_dm = df[['dropoff_longitude','dropoff_latitude']].reset_index(drop=True)
drop_location_dm['drop_location_dm_ind'] =drop_location_dm.index
```

```python
drop_location_dm =drop_location_dm[['drop_location_dm_ind','dropoff_longitude','dropoff_latitude']]
drop_location_dm.head()
```

```python
rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}
rate_code_dm = df[['RatecodeID']].reset_index(drop=True)
rate_code_dm['rate_code_id'] = rate_code_dm.index
rate_code_dm['rate_code_name'] = rate_code_dm['RatecodeID'].map(rate_code_type)


rate_code_dm = rate_code_dm[['rate_code_id','RatecodeID','rate_code_name']]
```

```python
rate_code_dm.head()
```

```python
passenger_count_dim = df[['passenger_count']].reset_index(drop=True)
passenger_count_dim['passenger_count_id'] = passenger_count_dim.index
```

```python
passenger_count_dim = passenger_count_dim[['passenger_count_id','passenger_count']]
passenger_count_dim.head()
```

```python
trip_distance_dim = df[['trip_distance']].reset_index(drop=True)
trip_distance_dim['trip_distance_id'] = trip_distance_dim.index
```

```python
trip_distance_dim = trip_distance_dim[['trip_distance_id','trip_distance']]
trip_distance_dim.head()
```

create a fact table  which are connect the multiple dimensonal table

```python
df = df.drop_duplicates().reset_index(drop=True)
df['trip_id'] = df.index
```

```python
fact_table = df.merge(datetime_dim,left_on='trip_id',right_on='datetime_id') \
               .merge(pickup_location_dm,left_on='trip_id',right_on='pickup_location_dm_index') \
               .merge(payment_type_dm,left_on='trip_id',right_on='payment_type_id') \
               .merge(drop_location_dm,left_on='trip_id',right_on='drop_location_dm_ind') \
               .merge(rate_code_dm,left_on='trip_id',right_on='rate_code_id') \
               .merge(trip_distance_dim,left_on='trip_id',right_on='trip_distance_id') \
               .merge(passenger_count_dim,left_on='trip_id',right_on='passenger_count_id') \
               [['trip_id','VendorID', 'datetime_id', 'passenger_count_id',
               'trip_distance_id', 'rate_code_id', 'store_and_fwd_flag', 'pickup_location_dm_index', 'drop_location_dm_ind',
                'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount',
               'improvement_surcharge', 'total_amount' ]]

```
